```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib notebook
```

```python
df = pd.read_csv('report_sokoban.csv')
df = df.drop(columns='Unnamed: 0', axis=1)
df['SearchType'] = df['SearchType'].str.replace('SearchType.', '')
df['SearchType'] = df['SearchType'].str.replace('_', ' ')
display(df)
```

```python
%matplotlib notebook
result = df.groupby(["SearchType"]).mean().reset_index().sort_values('Time')
display(result)
sns.barplot(x=result['SearchType'], y=result['Time']) #formerly: sns.barplot(x='Id', y="Speed", data=df, palette=colors, order=result['Id'])
plt.title('Mdia dos tempos de execuo para cada tipo de busca.')
sns.set(rc={'figure.figsize':(11.7,8.27)})
plt.show()
plt.savefig('graph1.png')
```

```python
%matplotlib notebook
result = df.groupby(["SearchType"]).mean().reset_index().sort_values('Node')
sns.barplot(x=result['SearchType'], y=result['Node']) #formerly: sns.barplot(x='Id', y="Speed", data=df, palette=colors, order=result['Id'])
plt.title('Mdia dos ns visitados para cada tipo de busca.')
sns.set(rc={'figure.figsize':(11.7,8.27)})
plt.show()
plt.savefig('graph2.png')
```
